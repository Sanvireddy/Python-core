``Lists``
- Used to group together values
- Might contain items of different data types
- Mutable (items can change their content)
- Simple assignment of a list to a variable never copies data. The variable refers to the existing list.
- Slicing operations creates a shallow copy
- Not all lists can be sorted (ex: lists containing different types of items)
- List comprehension - create a new list by applying a condition or expression to each item in an iterabe format
(typically, [(whatever inside append) for x in list if condition])
- Comparing 2 sequences - when one sequence is subset of other, then shorter one is the smaller


```python
list1 = ['Apple', 'Banana', 'Grapes']
list2 = list1
print(id(list1)==id(list2))
```

```python
list3 = [12,2]
list4= list3.sort()
print(list4) # None because it sorts inplace that means list3 now has sorted elements not l
print(list3)
list5 = [12,3,32]
list6 = sorted(list5)
print(list6) # returns a new list
```

| Method                          | Description                                                                                   |
|---------------------------------|-----------------------------------------------------------------------------------------------|
| `list.append(x)`                | Adds item `x` to the end of the list.                                                         |
| `list.extend(iterable)`         | Appends all items from the `iterable` to the list.                                            |
| `list.insert(i, x)`             | Inserts item `x` at position `i`.                                                             |
| `list.remove(x)`                | Removes the first item equal to `x`. Raises `ValueError` if not found.                       |
| `list.pop([i])`                 | Removes and returns item at index `i`. Defaults to last item. Raises `IndexError` if invalid.|
| `list.clear()`                  | Removes all items from the list.                                                              |
| `list.index(x[, start[, end]])`| Returns the index of the first item equal to `x`. Optional `start`/`end` limit the search.   |
| `list.count(x)`                 | Returns the number of times `x` appears in the list.                                          |
| `list.sort(*, key=None, reverse=False)` | Sorts the list in place. Optional `key` and `reverse` arguments for customization.     |
| `sorted(list)` | return a new sorted list |
| `list.reverse()`               | Reverses the list in place.                                                                   |
| `list.copy()`                   | Returns a shallow copy of the list. New list has copies of the references of elements of old list - not copies themselves |
| `copy.deepcopy(list)` | returns deep copy for mutable objects |


```python
import copy
p5 = list3.copy()
print(p5)
print(id(p5[0])==id(list3[0]))
p6 = copy.deepcopy(list3)
print(id(p6[0])==id(list3[0]))
'''
deepcopy() makes recursive copies of mutable objects.
For immutable ones (like numbers, strings, tuples), 
Python just reuses the same objects — hence the same id()
'''
```

```python
def summaryRanges(nums):
    start=nums[0]
    currstr=str(nums[0])
    outlist=[]
    for i in range(1,len(nums)): 
        if nums[i]!=start+1:

            finalstr=currstr+'->'+str(nums[i-1])
            outlist.append(finalstr)
            currstr=str(nums[i])
            start=nums[i]
    return outlist

les=summaryRanges([1,2,5,6,7,9])
print(*les)
```
