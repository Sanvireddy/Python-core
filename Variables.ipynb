- Variables are names pointing to value or objects stored in computer memory
- Python variables are `dynamically typed` that means variable types are determined and checked at `runtime` rather than during compilation, so it allows type changes through reassignment.
- Python variable can hold infinite value stored in finite computer memory
- In interactive mode (while running command line), the last printed expression is assigned to the variable `_`
- This `_` should not be used for reassignment as it creates a local independent variable with same name masking the built in variable with its magic behaviour
