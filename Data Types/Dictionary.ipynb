- Mapping objects maps hashable values to arbitrary objects
- If both keyword and positional arguments are provided, and a key appears in both, then keyword arguments value is considered

```python
a = dict(one=1, two=2, three=3)
b = {'one': 1, 'two': 2, 'three': 3}
c = dict(zip(['one', 'two', 'three'], [1, 2, 3]))
d = dict([('two', 2), ('one', 1), ('three', 3)])
e = dict({'three': 3, 'one': 1, 'two': 2})
f = dict({'one': 1, 'three': 3}, two=2)
a == b == c == d == e == f
```

| Method / Expression                          | Description                                                                                      |
|----------------------------------------------|--------------------------------------------------------------------------------------------------|
| `list(d)`                                     | Return a list of all the keys used in the dictionary `d`.                                       |
| `len(d)`                                      | Return the number of items in dictionary `d`.                                                    |
| `d[key]`                                      | Get the value for `key`. Raises `KeyError` if `key` is missing.                                 |
| `d[key] = value`                              | Set `d[key]` to `value`.                                                                         |
| `del d[key]`                                  | Delete `d[key]`. Raises `KeyError` if `key` is missing.                                          |
| `key in d`                                    | Return `True` if `key` is in the dictionary.                                                     |
| `key not in d`                                | Return `True` if `key` is not in the dictionary.                                                 |
| `iter(d)`                                     | Return an iterator over the dictionary’s keys.                                                   |
| `d.clear()`                                   | Remove all items from the dictionary.                                                            |
| `d.copy()`                                    | Return a shallow copy of the dictionary.                                                         |
| `dict.fromkeys(iterable, value=None)`         | Create a new dictionary with keys from `iterable` and values set to `value`.                     |
| `d.get(key, default=None)`                    | Return the value for `key`, or `default` if `key` is not found.                                  |
| `d.items()`                                   | Return a view of the dictionary’s items (key-value pairs).                                       |
| `d.keys()`                                    | Return a view of the dictionary’s keys.                                                          |
| `d.pop(key[, default])`                       | Remove `key` and return its value, or return `default` if `key` is missing.                      |
| `d.popitem()`                                 | Remove and return a (key, value) pair in LIFO order.                                             |
| `reversed(d)`                                 | Return a reverse iterator over the keys of the dictionary. (Added in Python 3.8)                 |
| `d.setdefault(key, default=None)`             | Return the value for `key`. If missing, insert `key` with `default` and return `default`.        |
| `d.update([other])`                           | Update dictionary with key/value pairs from `other`, overwriting existing keys.                  |
| `d.values()`                                  | Return a view of the dictionary’s values.                                                        |
