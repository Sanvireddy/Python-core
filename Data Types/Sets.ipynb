``Sets``
- unordered collection of elements
- set comprehensions are also supported
- cannot be used as a keys of dictionary

``Frozenset``
- Immutable - cannot edit after creation
- can be used as keys of a dictionary since its hashable

```python
basket = {'banana', 'apple', 'orange','apple'}
print(basket)

# empty set is created with set() and not {} as {} creates an empty dictionary
# set() is called type constructor
empty_set = set()
print(empty_set)    

a = set('banana')
print(a)
```

Below are applicable for `sets` and `frozenset`

| Expression / Method                     | Description                                                                                      |
|-----------------------------------------|--------------------------------------------------------------------------------------------------|
| `len(s)`                                | Return the number of elements in set `s` (cardinality of `s`).                                   |
| `x in s`                                | Test if `x` is a member of set `s`.                                                              |
| `x not in s`                            | Test if `x` is not a member of set `s`.                                                          |
| `s.isdisjoint(other)`                   | Return `True` if `s` and `other` have no elements in common.                                     |
| `s.issubset(other)` / `s <= other`      | Test whether every element in `s` is also in `other`.                                            |
| `s < other`                             | Test whether `s` is a **proper** subset of `other`.                                              |
| `s.issuperset(other)` / `s >= other`    | Test whether every element in `other` is also in `s`.                                            |
| `s > other`                             | Test whether `s` is a **proper** superset of `other`.                                            |
| `s.union(*others)` / `s | other | ...`  | Return a new set with elements from `s` and all others.                                          |
| `s.intersection(*others)` / `s & ...`   | Return a new set with elements common to `s` and all others.                                     |
| `s.difference(*others)` / `s - ...`     | Return a new set with elements in `s` but not in the others.                                     |
| `s.symmetric_difference(other)` / `s ^ other` | Return a new set with elements in either `s` or `other` but not both.                     |
| `s.copy()`                              | Return a shallow copy of set `s`.                                                                |


Only applicable to ``mutable sets``

| Method / Expression                                | Description                                                                                   |
|----------------------------------------------------|-----------------------------------------------------------------------------------------------|
| `s.update(*others)` / `s |= other | ...`             | Update the set, adding elements from all others.                                              |
| `s.intersection_update(*others)` / `s &= ...`       | Update the set, keeping only elements found in it and all others.                            |
| `s.difference_update(*others)` / `s -= ...`         | Update the set, removing elements found in others.                                            |
| `s.symmetric_difference_update(other)` / `s ^= ...` | Update the set, keeping elements in either set, but not in both.                             |
| `s.add(elem)`                                       | Add element `elem` to the set.                                                               |
| `s.remove(elem)`                                    | Remove element `elem` from the set. Raises `KeyError` if `elem` is not present.             |
| `s.discard(elem)`                                   | Remove element `elem` from the set if it is present. No error if it's missing.              |
| `s.pop()`                                           | Remove and return an arbitrary element. Raises `KeyError` if the set is empty.              |
| `s.clear()`                                         | Remove all elements from the set.                                                            |


```python
s1 = set('banana')
s2 = set('apple')
s1.update(s2)
print(s1)  
```
