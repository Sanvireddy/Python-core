| -------------- | -------------------------------------------------------------------------------- | --------------------------------------------------------------- |

| **Local**      | Created inside a function for parameters and local variables.                    | Created when function is called, destroyed when it returns.     |

````markdown
# 🧠 Notes on Namespaces, Memory Layout, and LEGB in Python

---

## 🧩 1. What is a Namespace?
A **namespace** is a container that holds names (identifiers) mapped to objects (values).  
It ensures that variable names don’t clash across different parts of the program.

Example:
```python
x = 10
print(x)
````

Here, `"x"` is a name stored in the **global namespace**, and it refers to the value `10`.

---

## 🗂️ 2. Types of Namespaces in Python

| Namespace Type | Description                                                                               | Lifetime                                 |
| -------------- | ----------------------------------------------------------------------------------------- | ---------------------------------------- |
| **Local**      | Created inside a function when it’s called. Holds variables defined within that function. | Until the function finishes execution    |
| **Enclosing**  | Created by outer functions in nested functions.                                           | Exists until the outer function finishes |
| **Global**     | Created when the module is loaded. Holds top-level variables, functions, and classes.     | Until the program ends                   |
| **Built-in**   | Contains Python’s built-in names like `len`, `print`, etc.                                | Always available                         |

---

## 🔁 3. The LEGB Rule (Scope Resolution)

When you use a variable, Python searches for it in this order:

1. **L – Local Namespace** → Inside the current function
2. **E – Enclosing Namespace** → In outer (enclosing) functions
3. **G – Global Namespace** → At the module level
4. **B – Built-in Namespace** → Predefined Python names

Example:

```python
x = 10  # Global

def outer():
    x = 20  # Enclosing
    def inner():
        x = 30  # Local
        print(x)
    inner()

outer()
```

Output:

```
30
```

Explanation:

* `x` inside `inner()` → **Local**
* If no local variable exists, Python searches **Enclosing**, then **Global**, then **Built-in**

---

## 🧱 4. Class and Instance Namespaces

Python classes and objects have their **own namespaces**.

| Namespace              | Description                           | Example        |
| ---------------------- | ------------------------------------- | -------------- |
| **Class Namespace**    | Contains class attributes and methods | `Car.wheels`   |
| **Instance Namespace** | Contains instance-specific variables  | `my_car.color` |

Example:

```python
class Car:
    wheels = 4  # Class namespace

    def __init__(self, color):
        self.color = color  # Instance namespace

my_car = Car("red")

print(Car.wheels)     # Class namespace
print(my_car.color)   # Instance namespace
```

**Important:**
Instance namespace is searched **before** the class namespace when accessing attributes.

---

## 💾 5. How RAM is Organized

RAM is divided into logical regions for program memory management:

| Memory Segment   | What it Stores                                             | Example                  |
| ---------------- | ---------------------------------------------------------- | ------------------------ |
| **Code/Text**    | Compiled machine code of the program                       | `def func(): ...`        |
| **Data Segment** | Global and static variables                                | `x = 10`                 |
| **Heap**         | Dynamically allocated memory (`objects`, `lists`, `dicts`) | `x = [1,2,3]`            |
| **Stack**        | Function call frames and local variables                   | Inside `def func(): ...` |

---

## 📍 6. Where Global Variables Live

Global variables in Python are **stored in the Data Segment** of RAM (conceptually).
They are **referenced in the Global Namespace** and remain alive until the program exits.

Example:

```python
x = 100  # Global variable → Global namespace, Data Segment
```

---

## 🔍 7. Relationship Between Heap and RAM

| Aspect         | RAM                       | Heap                               |
| -------------- | ------------------------- | ---------------------------------- |
| **Meaning**    | Physical memory           | A region inside RAM                |
| **Scope**      | Entire addressable memory | Used for dynamic allocations       |
| **Managed by** | OS                        | Runtime (Python memory manager)    |
| **Contains**   | Code, Stack, Data, Heap   | Only dynamically allocated objects |

**Memory trick:**
💡 “Heap lives **inside** RAM — it’s the playground where objects play freely!”

---

## ⚙️ 8. Quick Summary

| Concept                         | Description                                           |
| ------------------------------- | ----------------------------------------------------- |
| **Namespace**                   | A mapping between names and objects                   |
| **Scope**                       | Region of the program where a namespace is accessible |
| **LEGB**                        | Local → Enclosing → Global → Built-in                 |
| **Class vs Instance Namespace** | Class attributes shared; Instance attributes unique   |
| **Global Variables**            | Stored in Data Segment; accessed via Global Namespace |
| **Heap vs RAM**                 | Heap is part of RAM used for dynamic memory           |

>
> * **C** → Code
> * **D** → Data
> * **H** → Heap
> * **S** → Stack

```
RAM
 ├── Code Segment → function definitions, bytecode
 ├── Data Segment → global/static vars (global namespace)
 ├── Heap Segment → objects, lists, dicts
 └── Stack Segment → local vars, function calls
```

Namespaces follow:

```
Local → Enclosing → Global → Built-in
```
