``Tensors``
- Multidimensional array with uniform type (dtypes)
- Immutable like numbers and strings
- Convert to numpy array by using np.array or tensor.numpy
- Tf.tensors class allows tensors to be rectangular i.e., every element should be of same size along each axis. But below two types of tensors can have different shapes
- `Ragged tensors`: tensor with variable number of elements along the axis (this cannot be represented as regular tensor)
- `Sparse tensors`: 

- Tensors have shapes
    - `Shape` - Number of elements in each axis
    - `Rank` - Number of tensor axes
    - `Axis or Dimension` - Number of dimension of tensor
    - `Size` - Total number of items in the tensor

- `Broadcasting`: when there is a combined operation for large and small sized tensor, small sized tensor are streched automatically to fit larger tensors

```python
import tensorflow as tf

rank_1 = tf.ones([3,5,1])
print("Rank 1 Tensor:", rank_1[1,1,0].numpy())
```
