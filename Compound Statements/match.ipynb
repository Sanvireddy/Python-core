``match``

- `Guards` - if block beside case statement
    - It is execution if the `pattern in case block` succeeds, otherwise not evaluated
    - Pattern in case block succeeds, guard evalutes to true - case block is selected
    - Pattern in case block succeeds, guard evaluates to false - case block is not selected
    - Pattern in case block succeeds, guard throws an exception - exception bubbles up

- `or_pattern` `AS` `capture_pattern` - a concise way to match and bind
    - if it matches to the subject, then LHS(capture_pattern) is assigned the RHS(or_pattern) evaluated value
    - `_` is a dont care keyword so it cannot be used as capture_pattern

```python
match 42:
    case x:
        print(f"x= {x}")
```

``Wildcard pattern``
| **'_' usage** | **Meaning** |
| ------------- | ----------- |
|`_` in case | Wildcard - match anything, ignore it|
| case(_,x)| captures second value |
| `_`=10| Assigns value (Still a normal variable name) |

``Sequence Patterns``
- Used in match statements to destructure tuples or lists
- Subject (term beside match statement) should be tuples, lists, range but not string, bytes or bytearray
`Fixed Length Match` - 
- Subject must have the same number of elements as there are elements in pattern (not less, not more).

`Variable Length Match` 
- Subjects must have atleast the number of elements as there are non-star elements in the pattern 

```python
match([1,2,3,4,6]):
    case [a,b,c,d,e]:
        print(f"a= {a}, b= {b}, c= {c}, d= {d}, e= {e}")
    case [a,b,*rest]:
        print(f"a= {a}, b= {b}, rest= {rest}")
    case [a,*b,c]:
        print(f"a= {a}, b= {b}, c= {c}")
    case [a,*b]:
        print(f"a= {a}, b= {b}")
```

``Mapping Patterns``
- Used in match statements to destructure dict or any other class that implements Mapping
- Subject must be an instance of dict or any other type that implements Mapping
- All keys must be present in the pattern and subpattern must match
- Atmost one double star pattern is allowed and that must be at the end of the pattern
- No duplicates key are allowed


``Class Pattern``
- Matching class instances using class patterns

```python
match (1,2):
    case Point(x, y):
        print(f"Point x= {x}, y= {y}")

match 10:
    case int(0|1): # matches if exactly 0 or 1 (fails for 0.0 as it is not int)
        print("Matched 0 or 1")
```

- Point must be a class
- match checks isinstance(subject, Point)
- If positional arguments are used, python uses __match_args__ to map them to attributes
