``Functions ``
- Function definition is an executable statement
- Its execution binds the function name in the current local namespace with the function object. The current local namespace has the reference to global namespace

```python
# A list of type parameters can be given in square brackets after the function
# in version Python 3.12

def returnItem[T](items: T) -> T:
    return items[0]

print(returnItem('apple'))
print(returnItem([1,2,4]))
```

```python
# Default parameters can be given in the function definition
# Mutable default parameters should be avoided instead use None

def addItems(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items
print("Using None as default parameter")
print(addItems('apple'))
print(addItems('banana'))

def addItems2(item, items=[]):
    items.append(item)
    return items
print("Using mutable default parameters")
print(addItems2('apple'))
print(addItems2('banana'))
# The function can be called with a different type of item
```

|`` Symbol ``|`` Meaning ``|`` Use case ``|
|-------------|---------------|--------------|
|`/`| Before this -> Positional Arguments only | Hide Parameter names (lets you rename them later)|
|`*`| After this -> Keyword Arguments only | Forces clarity, prevents wrong argument order. Ex. parameter has length,width if no keywords while calling the function, unable to identify which is what|

```python
# Function Annotations can be used to specify the type of parameters and return type
def greet(name: str) -> str:
    return f"Hello, {name}!"
print(greet("Alice"))

# It is just giving us a hint that name should be a string and return type should be a string.
# Python does not enforce these annotations at runtime, but they are useful for documentation 
# and type checking tools. They are just metadata and do not affect the function's behavior.

print(greet(123))  # This will still work, but it's not recommended

# You can delay the evaluation of function annotations by importing __future__ annotations
# Without this, python evaluates annotations immediately
# This will help if the annotation is not defined and when circular references are present

p=[[4,15],[1,2],[4,5]]
p.sort()
print(p)
```

| Feature                         | Syntax Example                            | Description                                                                 |
|----------------------------------|--------------------------------------------|-----------------------------------------------------------------------------|
| Parameter annotation             | `x: int`                                   | Hints that parameter `x` should be of type `int`                            |
| Return type annotation           | `-> str`                                   | Hints that the function returns a `str`                                     |
| *args annotation                 | `*args: int`                               | Hints that `args` is a tuple of `int` values                                |
| **kwargs annotation              | `**kwargs: str`                            | Hints that kwargs is a dict of strings                                |
| Accessing annotations            | `func.__annotations__`                     | Returns a dictionary of all annotations                                     |                         |
| Special `*args` annotation       | `*args: *tuple[int, ...]`                  | Advanced use: variadic type hints (PEP 646)                                 |
| Postponed annotation evaluation  | `from __future__ import annotations`       | Keeps annotations as strings for delayed evaluation                         |
| Annotation evaluation timing     | At function definition                     | Annotations are evaluated when the function is **defined**, not when called |
