- Except clause specify one or more exception handlers

| **Situation**                              | **What Happens**                                                                 |
|-------------------------------------------|----------------------------------------------------------------------------------|
| `except` uses a valid exception class      | Python checks if it matches the raised exception (directly or via base class).  |
| `except` uses a tuple of exception classes | Python matches if any class in the tuple matches the raised exception.          |
| `except` expression raises an error        | Python stops handling the original exception and starts handling the new one.   |
| No matching `except` clause found          | Python looks up the call stack for another handler; if none, the program crashes.|
| Matching class is a base class             | It's accepted — the exception is caught if it inherits from the specified class.|



| **Concept**                        | **Explanation**                                                                 |
|-----------------------------------|---------------------------------------------------------------------------------|
| `as e` in `except` (except ZeroDivisionError as e)                | Binds the exception instance to variable `e`. (similar to e = ZeroDivisionError, actual exception is assigned to the variable e)                                  |
| Suite                             | The block under `except`, which is always executed if matched.                |
| After the suite                   | Execution resumes after the whole `try-except` block.                         |
| Nested `try-except` blocks        | Only the inner handler catches the error if it matches.                       |
| Clearing the exception variable   | After `except`, the bound exception (like `e`) is deleted automatically.      |


Exception is cleared after `except` 
- When we raise an exception python attaches a traceback (stact trace) to the exception object
- For ex- Exception -> traceback -> stacktrace -> local variables -> (back to) exception object
- If exception object (e) is not deleted, python will not let us cleaning up those frames and local variables
- Prevents reference cycles that cause memory leaks 


| **Concept**                     | **Explanation**                                                                 |
|--------------------------------|---------------------------------------------------------------------------------|
| Use another variable to keep it | If you want to use the exception after the block, assign it to another name   |
| `sys.exception()`              | Temporarily stores the current active exception inside an `except` block. Accessible inside the except block if it is not assigned      |
| `sys.exception()` is reset     | When leaving the handler, Python restores the previous exception (if any)     |


``except*``
- except* clause is used for handling Exception groups

| **Do's**                     | **Don'ts**                                                                 |
|--------------------------------|---------------------------------------------------------------------------------|
| except ExceptionName   |except ExceptionGroup |
| except* ExceptionName , except* ExceptionGroup | |
|    | Try block contain both except and except* |
| | Break, continue or return appear in except* clause|

``finally``
- Finally block does not contain the exception which is not raised in try, except clauses because it is not active in that scope
- But it is reraised after the finally block is executed (if finally block do no raise any exception)
