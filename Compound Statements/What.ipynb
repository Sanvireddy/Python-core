``Compound Statements``
- Group of statements that affect or control the execution of other statements
- Functions and class are syntactically compound statements
- A compound statement can consist multiple clauses 
    - a clause has a header and suite
    - a suite is a group of statement controlled by a clause
    - Each clause header starts with a uniquely identifying keyword and ends with a clause
- Names in target loop will not get deleted (i in for i in range() loop)

```python
for i in range(10):
    print("First Loop")
print(i)
for i in range(4):
    print("Second Loop")
print(i)
```
