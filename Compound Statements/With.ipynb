``with``
- It is used as a cleaner way to manage try/finally blocks
- It ensures that certain actions are executed when it enter/exits certain block of code


```python
with open('Functions.ipynb', 'r') as file:
    content = file.read()
    
# is similar to

file = open('Functions.ipynb', 'r')
try:
    content = file.read()
finally:
    file.close()
```
