collections.namedtuple(typename, field_names, *, rename=False, defaults=None, module=None)
- Returns a tuple sub class with name `typename`
- rename = True replaces all invalid field names with positional arguments. `_{index}`
- defaults = True will assign values from right to left. For example, (x,y,z) defaults to (1,2) that means x is required, y=1, z=2

```python
from collections import namedtuple

Point = namedtuple('Point', 'x,y,z',defaults=(1, 2))
pt = Point(4,5)
print(pt)
```

To prevent conflicts with field names, methods are defined starting with `_`

| **Method**        | **Description** |
|-------------------|---------------|
| **._make(iterable)**  | Makes a new instance from existing sequence of iterable. |
| **._asdict()**  | maps field names to values.|
| **._replace(kwargs)**  | replaces old values with a specified fields new value. |
| **._fields**  | Tuples of strings listing field names. |
| **._field_defaults()**  | Dictionary mapping fieldnames to default values. |
| **.getattr(typename, fieldname)**  | gets field whose value is stored in name. |

To convert dict to namedtuple, `**dict`

```python
print(pt._make('agk'))
print(pt._fields)
```
