- Counter is a dict subclass that counts hashable objects.
- It is a collection where elements are stored as dictionary keys and their count is stored as values
- Counter can be any integer(including zero or negative counts)
- Elements are counted using iterable (string or list) or initialised from another mapping (dictionary) or counter.
- Counter has a dictionary interface except that it returns count value to be zero for a element not found
- Modifying the element count to zero does not delete the element rather use `'del c[element]'` 

```python
from collections import Counter

iterable = Counter('aaabbbsssssscccccccc')
mapping = Counter({'a': 3, 'b': 2, 's': 6, 'c': 0})
keywordArgs = Counter(a=3, b=2, s=6, c=8)

print(iterable)
print(mapping)
print(keywordArgs)
```

- `Counter` is a subclass of `dict`, and since Python 3.7:
  - Dictionaries preserve insertion order.
  - `Counter` also preserves the order in which elements are first added.

- When performing **math operations** on `Counter` objects (like `+`, `-`, `&`, `|`):
  - The **result preserves order**.
  - The ordering rule is:
    1. Common elements appear first
    2. Elements appear in the order they were first seen in the **left operand**.
    3. Then elements unique to the **right operand** are appended in their original order.


```python
c1 = Counter({'d': 3,
              'b': 2,
              's': 6,
              'c': 8})
c2 = Counter({'s': 1,
                'a': 12,
                'b': 3,
                'c': 4})

c3 = c1 + c2
print(c3)
```

- Above output is misleading though the internal counter preserves the order
- Python uses __repr__ to print the string representation of object, this might do reordering of counter for printing purpose, but it preserves insertion order. 
Look below code for more details
- The elements() method requires integer counts. It ignores zero and negative counts.

```python
for key in c3:
    print(f'{key} : {c3[key]}')
```

| **Method**        | **Description** |
|-------------------|---------------|
| **.elements()**  | Returns an iterator over elements repeating each as many times as its count. |
| **.most_common(n)**  | Returns list of n most common and their counts from most common to least.|
| **.subtract([iterable or mapping])**  | Elements are subtracted from iterable or mapping. |
| **.total()**  | Sum of counts. |
| **.update()**  | It adds the count of elements rather than replacing it. |
| **.clear()**  | Resets count. |

```python
c = Counter(a=3, b=1)
d = Counter(a=1, b=2)
print(c + d )                      # add two counters together:  c[x] + d[x]

print(c - d)                    # subtract (keeping only positive counts)

print(c & d)                     # intersection:  min(c[x], d[x])

print(c | d)                    # union:  max(c[x], d[x])

print(c == d)                  # equality:  c[x] == d[x]
```
