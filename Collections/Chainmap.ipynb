- Chainmap groups dicts (mappings) to create a single updateable view.
- Underlying mappings are stored in a dict.
- First dict is treated as child and rest are parents
- Lookups search the underlying mappings successively until a key is found
- Writes/Updates/deletions happen for the first mapping

```python
from collections import ChainMap

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
cm = ChainMap(dict1, dict2)
print(cm)
```

```python
cm['b'] = 4
print(cm)

print(dict1)
# it stores the reference to the original dict, so it is modified
```

```python
cm.maps.append({'d': 5})
print(cm.maps)
```

| **Method**            | **Description** |
|-----------------------|-----------------|
| **.maps**        | user updatable list of mappings |
| **.new_child(m=None,kwargs)**    | Returns a new chainmap by creating a new map with this child followed by all other instances |
| **.parents**          | Returns a new chainmap returning all the instances except the first one. |

```python
user_preferences = {'theme': 'dark', 'language': 'en'}
default_preferences = {'theme': 'light', 'language': 'en', 'font_size': 12}
preferences = ChainMap(user_preferences, default_preferences)
print(preferences)
print(preferences['theme'])
print(preferences['font_size'])
```

- Chainmap only makes chanegs to the first mapping in the chain while searches for the whole tree
- Deep writes and insertions can be allowed by writing a subclass which does that
