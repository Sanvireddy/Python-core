- Deque (pronounced as double ended python queue)
- Reduces time complexity for prepending an element to O(1) due to its circular buffer
- Rotate method helps implement deque slicing and deletion

```python
from collections import deque

my_list = list(range(1_0))
my_deque = deque(range(1_0))

print(my_list)
print(my_deque)
```

```python
my_list.insert(10, None)

print(my_list)

my_deque.appendleft(10)
print(my_deque)
```

| **Method**            | **Description** |
|-----------------------|-----------------|
| **.append(x)**        | Add `x` to the right side of the deque. |
| **.appendleft(x)**    | Add `x` to the left side of the deque. |
| **.clear()**          | Remove all elements from the deque, leaving it empty. |
| **.copy()**           | Create a shallow copy of the deque. |
| **.count(x)**         | Count the number of elements equal to `x`.  |
| **.extend(iterable)** | Extend the right side by appending elements from the iterable. |
| **.extendleft(iterable)** | Extend the left side by appending elements from the iterable (in reverse order). |
| **.index(x[, start[, stop]])** | Return the index of `x` (optionally between `start` and `stop`). |
| **.insert(i, x)**     | Insert `x` at position `i`. |
| **.pop()**            | Remove and return the rightmost element. |
| **.popleft()**        | Remove and return the leftmost element. |
| **.remove(value)**    | Remove the first occurrence of `value`. |
| **.reverse()**        | Reverse elements in-place. |
| **.rotate(n=1)**      | Rotate the deque `n` steps to the right (left if `n` is negative). |
| **.maxlen** *(attribute)* | Maximum size of the deque or `None` if unbounded. |


```python
my_deque.extendleft([20, 30, 40])
print(my_deque)
```
