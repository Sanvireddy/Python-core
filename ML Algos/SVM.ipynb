### SVM using scikitlearn

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X,y = load_iris(as_frame=True, return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
plt.scatter(X_train[y_train == 0]['sepal length (cm)'],X_train[y_train == 0]['sepal width (cm)'],marker='+',color='r')
plt.scatter(X_train[y_train == 1]['sepal length (cm)'],X_train[y_train == 1]['sepal width (cm)'],marker='.',color='g')
plt.title("Class 0 vs Class 1")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal width (cm)")
```

```python
plt.scatter(X_train[y_train == 0]['petal length (cm)'],X_train[y_train == 0]['petal width (cm)'],marker='+',color='r')
plt.scatter(X_train[y_train == 1]['petal length (cm)'],X_train[y_train == 1]['petal width (cm)'],marker='.',color='g')
plt.title("Class 0 vs Class 1")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal width (cm)")
```

```python
model = SVC()
model.fit(X_train, y_train)
print(model.score(X_test,y_test))
```
